```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = r"939013_01.xls"
df = pd.read_excel(file)
df.head()
```

```python
df.drop(df.index[0:4], inplace=True) #excluir o cabeçalho antigo até a linha 4
df = df.reset_index(drop=True) #resetar os índices
df.columns = df.iloc[0] #definir a coluna de índice 0 como novo cabeçalho
df = df[1:] #Excluir a primeira linha, já que definimos o novo cabeçalho
df = df.reset_index(drop = True) #setar os índices
df.head()
```

```python
df[['ValFP', 'TipoFP']] = df['FPot.'].str.extract(r'(\d+)\s*([A-Za-z]+)')
df.head()
```

```python
df['ValFP'].fillna(0, inplace=True)
df['ValFP'] = df['ValFP'].astype(int)
df.head()
```

```python
df_Ind = df[df['TipoFP'] == 'L'][['ValFP', 'TipoFP']]
df_Ind = df_Ind.reset_index(drop=True)
display(df_Ind)
```

```python
df_Ind.describe()
```

```python
df_Ind_FP = df[(df['TipoFP'] == 'L') & (df['SH'] == 'F')][['ValFP', 'TipoFP','SH']]
df_Ind_FP = df_Ind_FP.reset_index(drop=True)
display(df_Ind_FP)
```

```python
df_Ind_FP.describe()
```

```python
df_Ind_P = df[(df['TipoFP'] == 'L') & (df['SH'] == 'P')][['ValFP', 'TipoFP','SH']]
df_Ind_P = df_Ind_P.reset_index(drop=True)
display(df_Ind_P)
```
